**GCA Measures:**

1. Participation
  * Mean participation of any participant above or below what you would expect from equal participation in a group of the size of theirs
2. Overall responsivity
  * Measure of how responsive a participant’s contributions are to all other group members’ recent contributions
3. Internal cohesion
  * How semantically similar a participant’s contributions are with the participant’s own recent contributions
4. Social impact
  * Measure of how contributions initiated by the corresponding participant have triggered follow-up responses
5. Newness
  * The amount of new information a participant provides, on average
6. Communication density
  * The amount of semantically meaningful information


```python
plt.figure(figsize=(16, 8))
heatmap = sns.heatmap(df3.corr(), annot = True, cmap='crest')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
```

**Cluster Names:**


1. Chatterers
  * May not always be supportive in online interactions
2. Drivers
  * Highest participators
  * High social impact, responsiveness, and internal cohesion
  * High degree of efforst in collaborative discussion
3. Followers
  * Lowest participation
  * Build on previous ideas
4. Lurkers
  * Lowest values across all GCA measures
5. Socially Detached
  * High internal cohestion, but low scores on other GCA values
  * Focused solely on themselves and their own narrative
6. Influential Actors
  * Low participation
  * When they contribute, they attend to other learners' contributions and provide meaningful information
  * High internal cohesion, overall responsiveness, and social impact



```python
plt.figure(figsize=(16, 8))
sns.scatterplot(data=df2, x= 'z.participation', y= 'z.newness', hue='SixCluster', palette="deep")
```

Cluster 1 and 2 have more participation.

```python
df4 = df[['z.participation',
 'z.social.impact',
 'z.overall.responsivity',
 'z.internal.cohesion',
 'z.newness',
 'z.comm.density',
 'SixCluster']]
```

```python
z_par = []
z_soc = []
z_overesp = []
z_internal = []
z_new = []
z_dens = []
for n in range(1,7,1):
  clus_n = df4.loc[df['SixCluster'] == n]
  z_par.append(sum(clus_n['z.participation']) / len(clus_n))
  z_soc.append(sum(clus_n['z.social.impact']) / len(clus_n))
  z_overesp.append(sum(clus_n['z.overall.responsivity']) / len(clus_n))
  z_internal.append(sum(clus_n['z.internal.cohesion']) / len(clus_n))
  z_new.append(sum(clus_n['z.newness']) / len(clus_n))
  z_dens.append(sum(clus_n['z.comm.density']) / len(clus_n))
```

```python
plotdata = pd.DataFrame({
    "z.participation":z_par,
    "z.social.impact":z_soc,
    "z.overall.responsivity":z_overesp,
    "z.internal.cohesion":z_internal,
    "z.newness":z_new,
    "z.comm.density":z_dens

    },
    index=["1", "2", "3", "4", "5", '6']
)
```

```python
plotdata.plot(kind="bar", figsize=(16, 8))
plt.title("GCA Measures for each Cluster")
plt.xlabel("Clusters")
plt.ylabel("Scores")
```
